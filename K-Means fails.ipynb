## - Where K-Means fails.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
print('success !!:)')
```

```python
X,Y = make_circles(n_samples=200,noise = 0.1)
X_,Y_ = make_moons(n_samples=200,noise = 0.1)
```

```python
print(X.shape,X_.shape)
```

# Linearily Inseperable data clusters:

### - Circle clusters

```python
plt.scatter(X[:,0],X[:,1],c = Y)
plt.show()
```

### - Moon clusters

```python
plt.scatter(X_[:,0],X_[:,1],c = Y_)
plt.show()
```

```python
kmeans = KMeans(n_clusters= 2)
kmeans.fit(X_)
```

```python
centers = kmeans.cluster_centers_
label = kmeans.labels_
```

```python
plt.scatter(X_[:,0],X_[:,1],c = label)
plt.scatter(centers[:,0],centers[:,1],c = 'black',marker = '*')
plt.show()
```

```python
kmeans.fit(X)
y_labels = kmeans.labels_
_centers = kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1],c = y_labels)
plt.scatter(_centers[:,0],_centers[:,1],marker = "*",color = "black")
plt.show()
```

## Conclusion : 
clearly, K-means algorithm fails for linearily inseperable data.
we use DBSCAN algorithm for such cases,for obtaining better result.
DBSCAN : density based spatial clustering on applications with noise.

```python
from sklearn.cluster import DBSCAN
```

```python
dbs = DBSCAN(eps = 0.2,min_samples=5)
```

```python
#working on moon data.
Ylabels = dbs.fit_predict(X_)
```

```python
plt.scatter(X_[:,0],X_[:,1],c = Ylabels)
plt.title('perfect clustering of moon dataset using DBSCAN')
plt.show()
```

```python
#perfect clustering is done using DBSCAN on moon dataset.
print(dbs.components_.shape)
```

```python
dbs2 = DBSCAN(eps = 0.12,min_samples = 5)
```

```python
#working on circle data set.
Y_labels = dbs2.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c = Y_labels)
plt.show()
```
